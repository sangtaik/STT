```python
import librosa
import numpy as np
import torch
import matplotlib.pyplot as plt
import sounddevice as sd
import soundfile as sf

sr = 16000

def plot_waveform(waveform, sample_rate, title="Waveform", xlim=None, ylim=None):
    
    num_frames = waveform.shape[0]
    time_axis = np.arange(num_frames) / sample_rate

    figure, axes = plt.subplots(1, 1)
    axes.plot(time_axis, waveform, linewidth=1)
    axes.grid(True)
    if xlim:
        axes.set_xlim(xlim)
    if ylim:
        axes.set_ylim(ylim)
    figure.suptitle(title)
    plt.show(block=False)
```

```python
speech, _ = librosa.load('./dataset/target.wav',sr)

fast_time_stretch = librosa.effects.time_stretch(speech,1.5)
slow_time_stretch = librosa.effects.time_stretch(speech,0.8)
```

```python
plot_waveform(speech, sr, title="Original", xlim=(0,5))
# sd.play(speech,sr)
```

```python
plot_waveform(fast_time_stretch, sr, title="fast", xlim=(0,5))
# sd.play(fast_time_stretch,sr)
sf.write('./outputs/target_fast.wav',fast_time_stretch,sr,'PCM_16')
```

```python
plot_waveform(slow_time_stretch, sr, title="slow", xlim=(0,5))
# sd.play(slow_time_stretch,sr)
sf.write('./outputs/target_slow.wav',slow_time_stretch,sr,'PCM_16')
```
